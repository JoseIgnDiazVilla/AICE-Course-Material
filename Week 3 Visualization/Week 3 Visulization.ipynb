### Figures and subplots

Any plot in Matplotlib is configured in an object called figure. This figure can be define implicitly by Matplotlib or explicitly by us and specify it's parameters. 

Let's begin with an empty figure with no other objects:

```python
fig = plt.figure()
plt.show()
```

We can also modify the size of the figure with the parameter figsize

```python
fig = plt.figure(figsize=(12,8))
plt.show()
```

We can see from the output that the figure has 0 Axes if we want to display any data we need to add some. 

There are many many ways to add Axes to a figure. From using a function like plot() to let Matplotlib handle it from high level to manually define and assign each one. 

As an example, let's add some Axes to a figure

```python
fig = plt.figure()
ax1 = fig.add_subplot() 
plt.show()
```

That's's a start, but we can also add multiple Axes to a single figure

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) 
ax2 = fig.add_subplot(2,2,2) 
ax3 = fig.add_subplot(2,2,3) 
plt.show()
```

The function plt.subplots() creates and initializes the Axes in the figure in a single line

```python
fig, axes = plt.subplots(2,3) # Here, axes it's a list with all the Axes objects
plt.show()
```

Now we can add some data to the Axes to display

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) 
ax2 = fig.add_subplot(2,2,2) 
ax3 = fig.add_subplot(2,2,3) 

ax1.plot(np.arange(10))
ax3.scatter(np.random.normal(size=100), np.random.normal(size=100))
plt.show()
```

```python
fig, axes = plt.subplots(3,3)

for i in range(3):
    for j in range(3):
        data = np.linspace(0, np.pi*2, 1000)
        x = np.cos(data*(i+1))
        y = np.sin(data*(j+1))
        axes[i, j].plot(x,y)
```

We can see some clipping of the figures with the axis, we can fix that by either increasing the figure size, reducing the number size or as we'll see now with the tight_layout() method.

```python
fig, axes = plt.subplots(3,3)

for i in range(3):
    for j in range(3):
        data = np.linspace(0, np.pi*2, 1000)
        x = np.cos(data*(i+1))
        y = np.sin(data*(j+1))
        axes[i, j].plot(x,y)

fig.tight_layout()
```


### Adding some style

Matplotlib also allows to modify how to display the data. Once again, there are many ways to achieve this, but let's go over some common ways to change the color, markers and styles in the plot. 

```python
x = np.arange(10)
y = np.random.normal(size=10)

plt.plot(x,y, color='g', linestyle='--') # g sets the color to green and -- sets the line to a dashed line
plt.show()
```

```python
x = np.arange(10)
y = np.random.normal(size=10)

plt.plot(x,y, 'g--') # the 'g--' it's a compressed argument for the color green (g) and the linestyle (--)
plt.show()
```

We can also add markers to the data points

```python
x = np.arange(10)
y = np.random.normal(size=10)

plt.plot(x,y, 'k*--') 
plt.show()
```

A guide for colors and linestyles can be found in the official Matplotlib documentation

### Axis

The axis (not to be confused by Axes) are the guides outside the canvas that indicates the reference for the data values. 

This are defined automatically with the creation of the Axes, but they can also be modified manually.

This can be done by using the methods set_xticks and set_xticklabels. It's important to remember that to use set_xlabels it's necessary to define the xticks first. 

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(500).cumsum())

ticks = ax.set_xticks([0, 36, 150, 225, 453])

labels = ax.set_xticklabels(['cero', 'A', 'B', 'X', '100'])

plt.show()
```

Analogically, the y labels can be modified by set_yticks and set_yticklabels

### Labels and Titles

When reading a graph, it's very important to know what you are looking at. Just displaying the data with no labels it's not only be useless, but detrimental. Luckily adding labels and even a title to any plot with Matplotlib it's quite simple.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Test Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

plt.show()
```

### Legends

Axes are not limited to a single data source to display. This allows to compare data directly and saves display space but without labeling the data it can get confusing like in the following example:

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(100).cumsum(), 'r')
ax.plot(np.random.randn(100).cumsum(), 'k--')
ax.plot(np.random.randn(100).cumsum(), 'g.-')
plt.show()
```

You would need to explain outside the plot which line corresponds to each data source. Instead, we can define a label to identify them directly on the plot.

```python
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(100).cumsum(), 'r', label='G_1')
ax.plot(np.random.randn(100).cumsum(), 'k--', label='G_2')
ax.plot(np.random.randn(100).cumsum(), 'g.-', label='G_3')

ax.legend() # Very important to call or the legend wont be displayed
plt.show()
```

# Why We Visualize

There are many books and whole university courses that analyze this question from a Data Science perspective. Due to the limited time we have let's try to reflect on this question by rambling for a bit.

#### A picture is worth a 1000 words

- Visualization makes your message simple to understand
- It articulates a clear understanding of a situation, with minimum distractions
- It helps identify emerging trends
- It shows relationships through patterns

#### Good visualization is more impactful

- People can remember your message
- You can deliver more information in less space
- People argue on words and numbers, less on figures
- It leverages the bias of “beauty is truth”

#### Tell a story

- Starts with a plot (e.g., a question)
- Tells a narrative (e.g, depicts a relationship)
- Has a clear end (e.g., an insight)

#### Stick to conventions

- Need to understand the message, not the structure (ej: time goes on the horizontal axis)

#### Has a purpose 

- Make sure your visual conveys the message
- Keep it simple. Don’t clutter your visual. Be selective

#### It's organized

- Emphasize the important
- Minimize the noise

## But then what makes a visualization 'bad'?

- It's overcomplicated
- Tries too many things at the same time
- Has no clear message
- Does not have a clear scale for the data
- Depends in hard concepts to visualize like angles or volumes 
- Does not consider the colors that could be associated to the data sources


# What is data preprocessing?
Data Preprocessing is a technique that is used to convert the raw data into a clean data set.
We collect data from a wide range of sources and most of the time, it is collected in raw format which is not applicable for the analysis.

***Quick fact***:  Research says that data scientists spend around 80% of their time only on preparing the data for processing and 76% of data scientists view data preparation as the least enjoyable part of their work’ ~ Forbes

## Why do we preprocess data?

Most of the time, we don’t get quality data. It often contains missing, noisy and inconsistent values. This can potentially reduce the accuracy of the end result. Therefore, we need to prepare data for processing further and escape from the bottleneck situation.


***quality data!  quality results!***

You can relate data preprocessing to the context of that story. Here, data is our tool, the axe! with which we are going to make our machine learn things/do the analysis to unravel so much of information. You need to clean and prepare that data before using it as a tool and get a surprising outcome.
