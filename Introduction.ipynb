## Welcome to the AICE Course for Data Science and AI Training

This course is designed to teach you the basics of Data Science and Machine Learning with a focus in research, self-study and practical applications.

The first weeks we will go over some basic concepts and tools that will be helpful in the later parts. 

This material is not designed as a university course, there are some assignments with feedback but no evaluations. How much you learn will depend on how you engage with the material.
